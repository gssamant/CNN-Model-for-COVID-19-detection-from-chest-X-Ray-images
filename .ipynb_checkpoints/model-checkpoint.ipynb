```python
import pandas as pd
import numpy as np
```

```python
urls = os.listdir('COVID-19_Radiography_Dataset/COVID/images')
```

```python
path = "COVID-19_Radiography_Dataset/COVID/images/" + urls[0]
```

```python
path
```

```python
def loadImages(path, urls, target):
  images = []
  labels = []
  for i in range(len(urls)):
    img_path = path + "/" + urls[i]
    img = cv2.imread(img_path)
    img = img / 255.0
    img = cv2.resize(img, (100, 100))
    images.append(img)
    labels.append(target)
  images = np.asarray(images)
  return images, labels
```

```python
covid_path = 'COVID-19_Radiography_Dataset/COVID/images'
covidurl = os.listdir(covid_path)
covidImages, covidTargets = loadImages(covid_path, covidurl, 1)
```

```python
len(covidurl), len(covidImages)
```

```python
normal_path = "COVID-19_Radiography_Dataset/Normal/images"
normal_urls = os.listdir(normal_path)
normalImages, normalTargets = loadImages(normal_path, normal_urls, 0)
```

```python
covidImages.shape
```

```python
normalImages.shape
```

```python
data = np.r_[covidImages, normalImages]
```

```python
data.shape
```

```python
targets = np.r_[covidTargets, normalTargets]
```

```python
targets.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, targets, test_size = 0.25)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential([
    Conv2D(32, 3, input_shape=(100, 100, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(16, 3, activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(16, 3, activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(optimizer=Adam(), loss=tf.keras.losses.BinaryCrossentropy(),metrics=['accuracy'])
```

```python
model.fit(x_train, y_train,batch_size=32,epochs=5,validation_data=(x_test, y_test), verbose = 2)
```

```python
plt.plot(model.history.history['accuracy'], label = 'train accuracy')
plt.plot(model.history.history['val_accuracy'],label = 'test_accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(model.history.history['loss'], label = 'train loss')
plt.plot(model.history.history['val_loss'],label = 'test_loss')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.compat.v1.losses import sparse_softmax_cross_entropy
import numpy as np
```

```python
img_path = 'COVID-19_Radiography_Dataset/Normal/images/Normal-1.png'
img = image.load_img(img_path, target_size=(100, 100))  
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0 

predictions = model.predict(img_array)

if predictions[0][0] > 0.5:
    print("The image is predicted to contain COVID-19.")
else:
    print("The image is predicted to be non-COVID.")
```
